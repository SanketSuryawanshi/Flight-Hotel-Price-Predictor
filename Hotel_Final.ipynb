```python
def get_soup(url):    
    
    '''
    Takes an url from Hotel.com and 
    infinitely scrolls down to end of page
    until no more content can be loaded.
    '''
    # Open up chrome in incognito
#     chrome_options = webdriver.ChromeOptions()
#     chrome_options.add_argument("--incognito")

    driver = webdriver.Chrome()

    # Nagivate to url 
    driver.get(url)
    
    # Scroll down until the end of the page
    has_scrolled_count = 0
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            
        for attempt in range(20):
            try:
                if driver.find_element_by_id("listings-loading").value_of_css_property("display") == "block":
                    has_scrolled_count = 0
                else:
                    time.sleep(0.05)
                    has_scrolled_count += 1
            except:
                continue

            
        if any([cur_elem.is_displayed() for cur_elem in driver.find_elements_by_class_name("info")]):
            break

        if has_scrolled_count > 20:
            break
            
    # Grabs the html of the fully scrolled-down page and parse it with BeautifulSoup  
    innerHTML = driver.execute_script('return document.body.innerHTML')
    soup = BeautifulSoup(innerHTML, 'lxml')
    driver.close()
    
#     driver.execute_script("return document.innerHtml;")
#     soup = BeautifulSoup(driver.page_source, "html.parser")
    
    return soup;
```

```python
def get_content_list(soup, tag, class_):
    raw = soup.find_all(tag, {'class': class_})
    raw_list = [content.text for content in raw]
    return raw_list    
```

```python
def combine_df_child(dates, adults = '2', children = '2'):
    
    combined_hotel_df = pd.DataFrame()
    
    for date in dates:
        url = generate_url_children(date[0], date[1], adults, children)
        soup = get_soup(url)
        partial_df = get_attributes(soup, date[0], date[1], adults, children)
        combined_hotel_df = pd.concat([combined_hotel_df, partial_df], ignore_index = True)
            
    print(f'This dataset has {combined_hotel_df.shape[0]} rows and {combined_hotel_df.shape[0]} columns')
    return combined_hotel_df 
```

#### Let's continue on with data preprocessing

We will extract datetime objects from the checkin and checkout dates. Here is how we can do this:

1. We need to extract the datetime objects from the checkin and checkout strings
2. Use the datetime object and calender.day_name attribute to determine the day of the week
3. Make two new columns for Checkin Day and Checkout Day

```python
hotel_df_uncleaned_datetime = hotel_df_uncleaned.copy()

# Extract the datetime object from checkin and checkout strings
checkin_datetime = [datetime.strptime(content, '%Y-%m-%d') for content in hotel_df_uncleaned['Checkin']]
checkout_datetime = [datetime.strptime(content, '%Y-%m-%d') for content in hotel_df_uncleaned['Checkout']]

Checkin_Day = [calendar.day_name[content.weekday()] for content in checkin_datetime]
Checkout_Day = [calendar.day_name[content.weekday()] for content in checkout_datetime] 

hotel_df_uncleaned_datetime['Checkin_Day'] = Checkin_Day
hotel_df_uncleaned_datetime['Checkout_Day'] = Checkout_Day
```

Let's put some context to these checkin and checkout days by categorizing it to be either:
* Friday through Sunday - Weekend
* Monday through Wednesday - Beginning of Week
* Wednesday through Friday - End of Week

```python
hotel_df_uncleaned_weekly_categories = hotel_df_uncleaned_datetime.copy()

# Initital an empty list 
When_in_Week = []
for day in Checkin_Day:
    if day == 'Friday':
        When_in_Week.append('Weekend')
    if day == 'Monday':
        When_in_Week.append('Beginning of Week')
    if day == 'Wednesday':
        When_in_Week.append('End of Week')

hotel_df_uncleaned_weekly_categories['When_in_Week'] = When_in_Week
```

```python
def distplot(df, feature, size):
    sns.distplot(df[feature])
    plt.xlabel(f'{feature}', size = size)
    plt.title(f'The distribution of {feature}', size = size)
```

```python
def find_feature_limit(df, feature):
    '''
    INPUT dataframe and feature
    Returns the lower and upper limit using IQR method
    '''
    
    q25, q50, q75 = np.percentile(df[feature], [25, 50, 75])
    IQR = q75 - q25
    
    # Calculate the max and min limits to be considered as an outlier
    
    lower = q25 - 1.5 * IQR
    upper = q75 + 1.5 * IQR
    
    return lower, upper
```

```python
def distplot_filter(df, feature, size):
    '''
    Plots the distribution of a numerical feature
    from df with a upper limit constraint defined by the
    feature's upper limit using the IQR method
    '''
    upper = find_feature_limit(df, feature)[1]
    distplot(df[df[feature] < upper], feature, size)
```

```python
def distplot_log(df, feature, size):  
    '''
    Plots the log distribution of a numberical feature
    from df
    '''
    sns.distplot(np.log(df[feature]))
    plt.xlabel(f'{feature}', size = size)
    plt.title(f'The log distribution of {feature}', size = size)
    
```

```python
def lm_cv(X, y):
    
    lm = LinearRegression()
    lm.fit(X, y)  
    y_prediction = lm.predict(X)
#     print(f'The Simple Regression R-Squared is {lm.score(X, y)}')
    
    kf = KFold(n_splits=5, shuffle=True, random_state = 101)
    R_Squared = np.mean(cross_val_score(lm, X_val, y_val, cv=kf, scoring='r2'))
    print(f'The R-squared using CV is {R_Squared}')
    MAE_e = MAE(np.exp(y), np.exp(y_prediction))
    RMSE_e = RMSE(np.exp(y), np.exp(y_prediction))
    print(f'The MAE is {MAE_e} and RMSE is {RMSE_e}')
    coef_list = list(zip(X_train.columns, lm.coef_))
    return (sorted(coef_list, key = lambda tuple: tuple[1], reverse = True), 
            R_Squared, MAE_e, RMSE_e, lm)
```

#### LassoCV

```python
def lasso_cv(X, y, alphavec, cv = 5):
    
    # Scalar Transformation
    std = StandardScaler()
    std.fit(X.values)
    X_train_scaled = std.transform(X.values)
    
    # Search the alpha vector and find the best alpha
    model_lassoCV = LassoCV(alphas = alphavec, max_iter = 10000, cv = cv)
    model_lassoCV.fit(X_train_scaled, y)
    y_prediction = model_lassoCV.predict(X_train_scaled)
    
    print(f'The optimal alpha is {model_lassoCV.alpha_}')
#     print(f'The R-Squared is {model_lassoCV.score(X_train_scaled, y)}')
    
    kf = KFold(n_splits = 5, shuffle = True, random_state = 101)
    R_Squared = np.mean(cross_val_score(Lasso(alpha = model_lassoCV.alpha_), X_train_scaled, y, cv = kf, scoring = 'r2'))
    print(f'The R-Squared value using CV is: {R_Squared}')
    MAE_e = MAE(np.exp(y), np.exp(y_prediction))
    RMSE_e = RMSE(np.exp(y), np.exp(y_prediction))
    print(f'The MAE is {MAE_e} and RMSE is {RMSE_e}')
    coef_list = list(zip(X.columns, model_lassoCV.coef_))
    return (sorted(coef_list, key = lambda tuple: tuple[1], reverse = True), 
            R_Squared, MAE_e, RMSE_e, model_lassoCV)    
```

#### RidgeCV

```python
def ridge_cv(X, y, alphavec, cv = 5):
    
    # Scalar Transformation
    std = StandardScaler()
    std.fit(X.values)
    X_train_scaled = std.transform(X.values)
    # Search the alpha vector and find the best alpha
    model_RidgeCV = RidgeCV(alphas = alphavec, cv = cv)
    model_RidgeCV.fit(X_train_scaled, y)
    y_prediction = model_RidgeCV.predict(X_train_scaled)
    
    print(f'The optimal alpha is {model_RidgeCV.alpha_}')
#     print(f'The RidgeCV Regression R-Squared is {model_RidgeCV.score(X_train_scaled, y)}')
    
    kf = KFold(n_splits=5, shuffle=True, random_state = 101)
    R_Squared = np.mean(cross_val_score(Ridge(alpha = model_RidgeCV.alpha_), X_train_scaled, y, cv=kf, scoring='r2'))
    print(f'The R_Squared using CV is {R_Squared}')
    MAE_e = MAE(np.exp(y), np.exp(y_prediction))
    RMSE_e = RMSE(np.exp(y), np.exp(y_prediction))
    print(f'The MAE is {MAE_e} and RMSE is {RMSE_e}')
    coef_list = list(zip(X.columns, model_RidgeCV.coef_))
    return (sorted(coef_list, key = lambda tuple: tuple[1], reverse = True), 
            R_Squared, MAE_e, RMSE_e, model_RidgeCV)
```

#### ElasticNetCV

```python
def ElasticNCV(X, y, alphavec, cv = 5):
    
    # Standard Scale our X features:
    std = StandardScaler()
    std.fit(X.values)
    X_train_scaled = std.transform(X.values)
    
    # Use ElasticNetCV to search for a range of alpha values setting l1_Ratio = 0.1
    model_ElasticNetCV = ElasticNetCV(alphas = alphavec, l1_ratio = 0.1, cv = cv, max_iter = 10000)
    model_ElasticNetCV.fit(X_train_scaled, y)
    y_prediction = model_ElasticNetCV.predict(X_train_scaled)
    
    print(f'The optimal alpha is {model_ElasticNetCV.alpha_}')
#     print(f'The ElasticNetCV Regression R-Squared is: {model_ElasticNetCV.score(X_train_scaled, y)}')

    kf = KFold(n_splits = 5, shuffle = True, random_state = 101)
    R_Squared = np.mean(cross_val_score(ElasticNet(alpha = model_ElasticNetCV.alpha_, l1_ratio = 0.1, max_iter = 10000), X_train_scaled, y, cv = kf, scoring = 'r2'))
    print(f'The R-Squared using CV is {R_Squared}')
    MAE_e = MAE(np.exp(y), np.exp(y_prediction))
    RMSE_e = RMSE(np.exp(y), np.exp(y_prediction))
    print(f'The MAE is {MAE_e} and RMSE is {RMSE_e}')
    coef_list = list(zip(X.columns, model_ElasticNetCV.coef_))
    return (sorted(coef_list, key = lambda tuple: tuple[1], reverse = False), 
            R_Squared, MAE_e, RMSE_e, model_ElasticNetCV)
```

```python
def plot_learning_curve(estimator, X, y, title):
    
    training_examples = []
    training_scores = []
    validation_scores = []
    sizes = [i for i in np.linspace(0.1,0.99,9)]
    
    for i in sizes:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i, random_state=42) 
        training_examples.append(X_train.shape[0])
        
        estimator.fit(X_train, y_train)
        train_pred = estimator.predict(X_train)
        training_score = MAE(np.exp(y_train), np.exp(train_pred))
        training_scores.append(training_score)
        

        test_pred = estimator.predict(X_test)
        validation_score = MAE(np.exp(y_test), np.exp(test_pred))
        validation_scores.append(validation_score)
    
    plt.figure(figsize = (18,8));
    sns.lineplot(training_examples, training_scores, label="Training");
    sns.lineplot(training_examples, validation_scores, label="Cross-validation score");
    
    plt.xlabel("Training examples", size = 15);
    plt.ylabel("Mean Absolute Error", size = 15);
    plt.title(title, size = 20);
    plt.ylim([29, 35]);
    plt.xlim([0, 12000]);
#     plt.legend(loc="best");
    plt.legend(fontsize='x-large', title_fontsize='100', loc = 'best');
    return plt
    
```
